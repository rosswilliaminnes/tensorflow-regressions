```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# need to add column names

cols = ['CRIM','Zone','INDUS','CHAL','NOX con','Rooms','Age',
        'Dis','RAD','TAX','Pupil Ratio','Black','LSTA','Median val']

dataset.columns = cols
```

```python
# create dep var and feture matrix, the dep var is 'Media Val'

X=dataset.iloc[:,0:13]
y=dataset.iloc[:,13].values
```

no categorical features so no need to one hot encode, features are all numerical so lets scale them!
min max scalar is good if data is normally distributed

```python
# scale both the features and target (or dep variable)
from sklearn.preprocessing import  MinMaxScaler
sc= MinMaxScaler()
X= sc.fit_transform(X)
y= y.reshape(-1,1) # get into correct format so can sc.fit_transform
y=sc.fit_transform(y)
```

```python
# create test and training set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```
